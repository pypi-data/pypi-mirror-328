## Compare LIMEPY to Plummer

The LIMEPY models are isothermal in the inner parts, and behave like polytropes in the outer parts. In the limit of the central potential going to 0 we thus obtain polytropic models. The model with $g=3.5$ has a polytropic index of $n=5$, just as the famous Plummer (1911) model. Below we show that for $\lim_{\phi_0\rightarrow0}$ and $\lim_{g\rightarrow3.5}$ the LIMEPY model is similar to Plummer's model

```python
%matplotlib inline
from limepy import limepy
import matplotlib.pyplot as plt
import numpy 

r_pl = numpy.logspace(-1,5,100)
# Density for Plummer (1911) model with G=M=r0=1
rho_pl = 0.75/numpy.pi*(1+r_pl**2)**(-2.5)

# LIMEPY model scaled to same units. Note: rv = 16/(3*pi*r0)
l = limepy(0.001, 3.499, M=1,G=1, rv=16./(3*numpy.pi))

plt.plot(r_pl, rho_pl)
plt.plot(l.r, l.rho)
plt.loglog()

```

(Positive) potentials

```python
phi_pl = (1+r_pl**2)**(-0.5)
plt.loglog()
plt.plot(r_pl, phi_pl)
plt.plot(l.r, l.phi)
```
