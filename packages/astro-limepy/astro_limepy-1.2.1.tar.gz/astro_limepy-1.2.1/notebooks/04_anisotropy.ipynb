## Anisotropy

One of the advantages of LIMEPY is that it is straightforward to include radially biased anisotropy. The results are quickly obtained because (quasi-)analytic expressions for the density as a function of potential and radius are derived in Gieles & Zocchi 2015. Here is how to generate a King (1966) model with anisotropy (often referred to as Michie-King model) with an anistropy radius of 7 King radii

```python
from limepy import limepy
a = limepy(7,1,ra=7)
```

The anisotropy profile $\beta=1-\frac{\sigma_{\rm t}^2}{2\sigma_{\rm r}^2}$ looks like this

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(a.r, a.beta)
```

The effect on the density of adding radial anisotropy is similar to increasing the truncation parameter $g$

```python
k = limepy(7,1)
w = limepy(7,1.7)
plt.loglog()
plt.axis([1e-1, 1e3, 1e-10, 2])
plt.plot(k.r, k.rho,label='Isotropic | g=1')
plt.plot(a.r, a.rho,label='Michie-King | g=1')
plt.plot(w.r, w.rho,label='Isotropic | g=1.7')
plt.legend()
```
